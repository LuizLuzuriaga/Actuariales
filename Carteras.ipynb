# TEMA 1. Teoria de carteras analisis media varianza. Frontiera eficiente
La teoría de carteras y el análisis media-varianza son conceptos fundamentales en finanzas que ayudan a los inversores a optimizar sus carteras de inversión. A continuación, se presenta una explicación básica de estos conceptos y la frontera eficiente.
## Teoría de Carteras
La teoría de carteras, desarrollada por Harry Markowitz en la década de 1950, se centra en la idea de que los inversores pueden construir carteras que maximicen el rendimiento esperado para un nivel dado de riesgo, o minimicen el riesgo para un nivel dado de rendimiento esperado. La diversificación es un componente clave de esta teoría, ya que al combinar diferentes activos, se puede reducir el riesgo total de la cartera.
## Análisis Media-Varianza
El análisis media-varianza es una técnica utilizada para evaluar y optimizar carteras de inversión. Se basa en dos métricas principales:
- **Media (Rendimiento Esperado)**: Es la media ponderada de los rendimientos esperados de los activos en la cartera.
- **Varianza (Riesgo)**: Es una medida de la dispersión de los rendimientos de los activos en la cartera. La varianza se utiliza para cuantificar el riesgo asociado con la cartera.

El objetivo del análisis media-varianza es encontrar la combinación óptima de activos que maximice el rendimiento esperado para un nivel dado de riesgo o minimice el riesgo para un nivel dado de rendimiento esperado.
## Frontera Eficiente   
La frontera eficiente es un concepto clave en la teoría de carteras que representa el conjunto de carteras óptimas que ofrecen el mayor rendimiento esperado para un nivel dado de riesgo o el menor riesgo para un nivel dado de rendimiento esperado. En un gráfico donde el eje X representa el riesgo (desviación estándar) y el eje Y representa el rendimiento esperado, la frontera eficiente se muestra como una curva ascendente. Las carteras que se encuentran en la frontera eficiente son consideradas óptimas, mientras que las carteras que se encuentran por debajo de la frontera son subóptimas, ya que ofrecen un menor rendimiento para el mismo nivel de riesgo.
### Supuestos de Harry Markowitz
1. Hay una serie de activos financieros A_1, A_2, ..., A_n disponibles para invertir. Normalmente son acciones y bonos.
2. Disponemos de un capital inicial C_0 que podemos invertir en los activos financieros. Pero el problema principal es como distribuir el capital entre los distintos activos.
3. Asi, se consideran carteras x = (x_1, x_2, ..., x_n) donde x_i es el porcenteja de la cantidad de dinero invertida en el activo A_i. Por tanto, se tiene que cumplir que:
   - x_i >= 0 para todo i = 1, 2, ..., n (no se permiten ventas en corto)
   - x_1 + x_2 + ... + x_n = 1 (la suma de los porcentajes debe ser igual a 100%)
4. Cada activo A_i tiene un rendimiento aleatorio R_i, que se puede modelar como una variable aleatoria con una distribución de probabilidad conocida.
5. Se asume que los rendimientos de los activos son independientes entre sí.
6. El inversor es racional y busca maximizar su utilidad esperada, lo que implica que prefiere carteras con mayor rendimiento esperado y menor riesgo.

- Markowitz, H. (1952). Portfolio Selection. The Journal of Finance, 7(1), 77-91.
- Bodie, Z., Kane, A., & Marcus, A. J. (2014). Investments (10th ed.). McGraw-Hill Education.

### Rentabilidad esperada y varianza de una cartera de valores: Matriz de varianzas-covarianzas
La rentabilidad esperada de una cartera de valores se calcula como la suma ponderada de las rentabilidades esperadas de los activos individuales que componen la cartera. Si tenemos una cartera con n activos, la rentabilidad esperada \( E(R_p) \) de la cartera se puede expresar como:
\[ E(R_p) = \sum_{i=1}^{n} x_i E(R_i) \]
donde:   
- \( x_i \) es el peso del activo i en la cartera (la proporción del total invertido en ese activo).
- \( E(R_i) \) es la rentabilidad esperada del activo i.

La varianza de la rentabilidad de la cartera, que mide el riesgo asociado a la cartera, se calcula utilizando la matriz de varianzas-covarianzas de los activos. La varianza \( \sigma_p^2 \) de la cartera se puede expresar como:
\[ \sigma_p^2 = \sum_{i=1}^{n} \sum_{j=1}^{n} x_i x_j \sigma_{ij} \]
donde:
- \( \sigma_{ij} \) es la covarianza entre los rendimientos de los activos i y j. Si i = j, entonces \( \sigma_{ii} \) es la varianza del activo i.

Entonces cual es la rentabilidad de una cartera para responder se necesita hablar de rentabilidad de cada activo. La rentabilidad de cada activo se puede calcular como:
\[ R_i = \frac{P_{i,t} - P_{i,t-1} + D_i}{P_{i,t-1}} \]
donde:
- \( P_{i,t} \) es el precio del activo i en el tiempo t. Depende del horizonte de plantificación del inversor.
- \( P_{i,t-1} \) es el precio del activo i en el tiempo t-1. 
- \( D_i \) es el dividendo pagado por el activo i durante el período

El precio inicial del activo en t-1 es conocido, pero el precio en t es incierto. Por tanto, la rentabilidad \( R_i \) es una variable aleatoria. La rentabilidad esperada \( E(R_i) \) se puede estimar utilizando datos históricos o mediante modelos financieros. 

Si suponemos que los dividendos se reinvierten en el mismo activo, la fórmula de la rentabilidad se simplifica a:
\[ R_i = \frac{P_{i,t}}{P_{i,t-1}} - 1 \]
En resumen, para calcular la rentabilidad esperada y la varianza de una cartera de valores, se necesita conocer las rentabilidades esperadas de los activos individuales, sus pesos en la cartera y la matriz de varianzas-covarianzas que describe las relaciones entre los rendimientos de los activos.

La rentabilidad de la cartera se puede deducir por medio de dos pasos:
1. Si en un momento inicial tenemos un capital C_0 en t o en el momento final (f) tenemos un capital C_f:
\[C_f = C_0 (1 + R_p)\]
2. Despejando R_p:
\[R_p = \frac{C_f - C_0}{C_0} = \frac{C_f}{C_0} - 1\
Donde R_p es la rentabilidad de la cartera.
]

Si invertimos en una cartera en varios activos con diferentes ponderaciones x_i, la rentabilidad de la cartera se puede expresar como:
\[R_p = \sum_{i=1}^{n} x_i R_i\]
Donde R_i es la rentabilidad del activo i.

Ya que cada R_i es una variable aleatoria, R_p también lo es. Por tanto, se puede calcular la rentabilidad esperada de la cartera como:
\[E(R_p) = \sum_{i=1}^{n} x_i E(R_i)\]
Donde E(R_i) es la rentabilidad esperada del activo i. La \(R_p\) no se puede maxmizar al ser una variable aleatoria, pero si se puede maximizar su rentabilidad esperada \(E(R_p)\). Markowitz propone maximizar \(E(R_p)\) para un nivel de riesgo dado o minimizar el riesgo para un nivel de rentabilidad esperada dado.

Para medir el riesgo propone la desviación típica. Supone que las \(r_i\) tienen distribuciones simétricas, y entonces demuestra que minimizando la desviación típica, se maximiza la utilidad del inversor (Principio de Dominancia Estocástica de Segundo Orden).

La realidad es que las rentabilidades de los activos no siguen distribuciones simetricas sino asimetricas. Como ejemplo los derivados financieros como futuros, forwards u opciones financieras. Por lo que la varianza no sirve para medir el riesgo. Es mucho más factible el CVaR (Conditional Value at Risk) o el VaR (Value at Risk) u otras medidas coherentes y acotadas por la media.

Entonces, cual es el riesgo de una cartera de activos x = (x_1, x_2, ..., x_n)? La varianza de la cartera se puede calcular como:
\[\sigma_p^2 = \sigma_{11} x_1^2 + \sigma_{22} x_2^2 + ... + \sigma_{nn} x_n^2 + 2 \sigma_{12} x_1 x_2 + 2 \sigma_{13} x_1 x_3 + ... + 2 \sigma_{1n} x_{1} x_n + 2 \sigma_{23} x_2 x_3 + ... + 2 \sigma_{2n} x_2 x_n + ... + 2 \sigma_{n-1,n} x_{n-1} x_n\]
Donde \(\sigma_{ij}\) es la covarianza entre los activos i y j. Si i = j, entonces \(\sigma_{ii}\) es la varianza del activo i.

Otra expresion es por medio del coeficiente de correlación \(\rho_{ij}\):
\[\sigma_p^2 = \sigma_1^2 x_1^2 + \sigma_2^2 x_2^2 + ... + \sigma_n^2 x_n^2 + 2 \rho_{12} \sigma_1 \sigma_2 x_1 x_2 + 2 \rho_{13} \sigma_1 \sigma_3 x_1 x_3 + ... + 2 \rho_{1n} \sigma_1 \sigma_n x_{1} x_n + 2 \rho_{23} \sigma_2 \sigma_3 x_2 x_3 + ... + 2 \rho_{2n} \sigma_2 \sigma_n x_2 x_n + ... + 2 \rho_{n-1,n} \sigma_{n-1} \sigma_n x_{n-1} x_n\]
Donde \(\rho_{ij}\) es el coeficiente de correlación entre los activos i y j, y \(\sigma_i\) es la desviación típica del activo i.

Como a cada activo le corresponde una varianza y una desviación típica, se puede construir la matriz de varianzas-covarianzas:
\[\Sigma = \begin{pmatrix}
\sigma_{11} & \sigma_{12} & ... & \sigma_{1n} \\
\sigma_{21} & \sigma_{22} & ... & \sigma_{2n} \\
... & ... & ... & ... \\         
\sigma_{n1} & \sigma_{n2} & ... & \sigma_{nn} \\
\end{pmatrix}\]
Donde \(\sigma_{ij}\) es la covarianza entre los activos i y j. Si i = j, entonces \(\sigma_{ii}\) es la varianza del activo i. Esta matriz es semidefinida positiva y simetrica. 

Entonces, la varianza de la cartera se puede expresar de manera matricial como:
\[\sigma = x^T \Sigma x\]
Donde \(x\) es el vector de ponderaciones de la cartera, \(\Sigma\) es la matriz de varianzas-covarianzas, y \(x^T\) es la transpuesta del vector \(x\). 
Esta expresión permite calcular la varianza de la cartera de manera eficiente utilizando álgebra matricial. Además \(\rho_{ij} = \frac{\sigma_{ij}}{\sigma_i \sigma_j}\).

Sabiendo esto se puede plantear el problema de optimización de la cartera como:
\[\min_{x} x^T \Sigma x\]
Sujeto a:
\[\sum_{i=1}^{n} x_i E(R_i) = R_t\]
\[\sum_{i=1}^{n} x_i = 1\]
Donde \(R_t\) es el rendimiento esperado objetivo de la cartera. Este problema busca minimizar el riesgo (varianza) de la cartera para un nivel dado de rendimiento esperado.

````python
# se tienen las siguientes estrategias
# Estrategia X: 1. 50% de probabilidad de ganar 1.5
# Estrategia X: 2. 25% de probabilidad de ganar 3.5
# Estrategia X: 3. 25% de probabilidad de ganar 4.5

# Estrategia Y: 1. 50% de probabilidad de ganar 1.5
# Estrategia Y: 2. 50% de probabilidad de ganar 4.3

# Comprueba si una de ellas domina a la otra desde el punto de vista de algunas de las formas estudiadas (Dominancia, DEPO, DESO)

#Dominancia Plena es aquella que supera a una de las carteras. X domina plenamente a Y si en todos los escenarios posibles X tiene un rendimiento mayor o igual que Y y en al menos uno de los escenarios X tiene un rendimiento estrictamente mayor que Y.?
import numpy as np
from scipy.stats import entropy
#Usamos numpy y scipy para cálculos estadísticos. Entropy no es necesario aquí, pero se importa por si se requiere en análisis más avanzados.
#def en python funciona como una definición de función.

# Definimos las estrategias
estrategia_X = [(0.5, 1.5), (0.25, 3.5), (0.25, 4.5)]
estrategia_Y = [(0.5, 1.5), (0.5, 4.3)]
# Función para calcular el valor esperado
def valor_esperado(estrategia):
    return sum(prob * ganancia for prob, ganancia in estrategia)
# Función para calcular la dominancia plena
def dominancia_plena(estrategia_A, estrategia_B):
    ganancias_A = [ganancia for prob, ganancia in estrategia_A]
    ganancias_B = [ganancia for prob, ganancia in estrategia_B]
    domina = all(a >= b for a, b in zip(ganancias_A, ganancias_B)) and any(a > b for a, b in zip(ganancias_A, ganancias_B))
    return domina
# Calcular valores esperados
ve_X = valor_esperado(estrategia_X)
ve_Y = valor_esperado(estrategia_Y)
print(f"Valor Esperado Estrategia X: {ve_X}")
print(f"Valor Esperado Estrategia Y: {ve_Y}")
# Comprobar dominancia plena
if dominancia_plena(estrategia_X, estrategia_Y):
    print("La Estrategia X domina plenamente a la Estrategia Y")
else:
    print("La Estrategia X no domina plenamente a la Estrategia Y")

#Vamos con la Dominancia Estocástica de Primer Orden (DEPO)
def _cdf_en_puntos(estrategia, puntos):
    """
    Devuelve F(x)=P(G<=x) evaluada en 'puntos' para una estrategia discreta [(p, g), ...].
    """
    probs = np.array([p for p, _ in estrategia], dtype=float)
    gains = np.array([g for _, g in estrategia], dtype=float)

    # F(x) = sum p_i * 1{g_i <= x}
    return np.array([(probs[gains <= x]).sum() for x in puntos], dtype=float)

def depo(estrategia_A, estrategia_B, tol=1e-12):
    """
    A domina a B en DEPO (FSD) si su CDF es <= en todos los puntos (y < en alguno).
    Se evalúa sobre el soporte conjunto (todas las ganancias posibles).
    """
    soporte = sorted(set([g for _, g in estrategia_A] + [g for _, g in estrategia_B]))
    cdf_A = _cdf_en_puntos(estrategia_A, soporte)
    cdf_B = _cdf_en_puntos(estrategia_B, soporte)

    domina_todo = np.all(cdf_A <= cdf_B + tol)
    domina_estricto = np.any(cdf_A < cdf_B - tol)
    return bool(domina_todo and domina_estricto)

# Comprobar DEPO
if depo(estrategia_X, estrategia_Y):
    print("La Estrategia X domina a la Estrategia Y en DEPO")
else:
    print("La Estrategia X no domina a la Estrategia Y en DEPO")
````

##Value at Risk (VaR) y Conditional Value at Risk (CVaR)
El Value at Risk (VaR) y el Conditional Value at Risk (CVaR) son dos medidas de riesgo ampliamente utilizadas en finanzas para cuantificar el riesgo de pérdida en una cartera de inversión.

Hasta ahora hemos venido trbajando con modelos de renta variable para rendimientos simetricos. Pero en la realidad los rendimientos no son simetricos, sino asimetricos o presentan problemas con colas pesadas. Por ejemplo, los derivados financieros como futuros, forwards u opciones financieras. Por lo que la varianza no sirve para medir el riesgo. 

De hecho no debemos usar la desviacion tipica como medida de riesgo por no ser compatible con la teoria de la utilidad y la dominancia estocastica de segundo orden. Es mucho más factible el CVaR (Conditional Value at Risk) o el VaR (Value at Risk) u otras medidas coherentes y acotadas por la media.

La forma de medir y gestionar el riesgo dependera no solo del tipo de activo sino tambien de la clase de riesgo que se quiera medir (riesgo de mercado, riesgo de credito, riesgo de liquidez, riesgo operacional, etc.)

A modo de apartado historico,  a lo largo del siglo pasado han ocurrido episodios de corrupcion financiera y crisis economicas que han puesto de manifiesto la importancia de medir y gestionar el riesgo en las instituciones financieras. El 26 de Junio de 1974 el sitema aleman forzo la liquidacion del banco Herstatt en Alemania, esto acelero el comite de Basilea a publicar en 1988 el primer acuerdo de capital (Basel I) para regular el capital minimo que debian mantener los bancos. Posteriormente, en 2004 se publico Basel II y en 2010 Basel III, que introdujeron nuevas medidas para gestionar el riesgo de credito, liquidez y operacional. Existen otros marcos regulatorios como Solvencia II para las aseguradoras y MiFID II para los mercados financieros en Europa. Solvencia II consiste en un marco regulatorio para las aseguradoras que establece requisitos de capital basados en el riesgo y promueve una gestión prudente del riesgo y se basa en tres pilares: requisitos cuantitativos, gobernanza y supervisión. En la practica, Solvencia II se sigue de manera que las aseguradoras deben calcular su capital requerido utilizando modelos internos o estándares, y deben mantener un capital suficiente para cubrir sus riesgos.

Así los requerimientos de capital parecen ser una buena alternativa a la desviacion tipica. En efecto, miden el riesgo en terminos monetarios lo que es una gran novedad. Ademas una entidad o una cartera de valores  que refleje un alto nivel de riesgo debera aportar altas reservas iniciales como requerimientos de capital mientras que esas reservas seran bajas si el nivel de riesgo es bajo.

Asi esta nueva forma de medir parece resolver los dos problemas citados al comienzo: Pudiendo ser aplicable a cualquier tipo de inversion y de riesgo y permitiendo a reguladores y supervisores mantener el control de gestores y entidades mediante la aplicaion nde nirmas legales y escritas lo que defiende al inversor  no cualificado y al sistema en general incrementando la confianza del mismo. 

Entonces la pregunta que surje es como podemos fijar el sistema de requerimientos de capital? Una forma es mediante el Value at Risk (VaR) y el Conditional Value at Risk (CVaR).
## Value at Risk (VaR)
El Value at Risk (VaR) es una medida estadística que cuantifica la pérdida máxima esperada en una cartera de inversión durante un período de tiempo específico, con un nivel de confianza determinado. En otras palabras, el VaR responde a la pregunta: "¿Cuál es la pérdida máxima que puedo esperar en mi cartera con un nivel de confianza del X% durante el próximo día/semana/mes?"

No es más que un percentil de una distribucion de probabilidad. Una vez fijado el nivel de confianza, y representa la minima perdida posible al nivel de confianza dado. 

Se expresa como para un VaR al 95% es de 10000 euros  se quiere decir que si se cae en el peor 5% de los casos, la perdida sera al menos de 10000 euros. Será de 10000 euros o más. 

Inconveneintes del VaR:
- No es una medida coherente de riesgo: No cumple con todas las propiedades deseables de una medida de riesgo coherente, como la subaditividad.
- No proporciona información sobre la magnitud de las pérdidas más allá del umbral del VaR: El VaR solo indica la pérdida máxima esperada hasta un cierto nivel de confianza, pero no proporciona información sobre las pérdidas que pueden ocurrir en el peor escenario.
- Sensible a la distribución de los rendimientos: El VaR puede ser sensible a la forma de la distribución de los rendimientos, lo que puede llevar a subestimar el riesgo en presencia de colas pesadas o eventos extremos.
## Conditional Value at Risk (CVaR)
El Conditional Value at Risk (CVaR), también conocido como Expected Shortfall, es una medida de riesgo que complementa al VaR al proporcionar información sobre la magnitud de las pérdidas en el peor escenario. El CVaR se define como la pérdida promedio esperada en los casos en los que la pérdida excede el VaR. En otras palabras, el CVaR responde a la pregunta: "¿Cuál es la pérdida promedio esperada en mi cartera en el peor X% de los casos?"
El CVaR es una medida coherente de riesgo, ya que cumple con todas las propiedades deseables de una medida de riesgo coherente, incluyendo la subaditividad. Además, el CVaR proporciona información adicional sobre la magnitud de las pérdidas en el peor escenario, lo que lo hace más útil para la gestión del riesgo en comparación con el VaR.

### Cálculo del VaR y CVaR
En este repositorio se calculara el VaR mediante el metodo parametrico. Aunque existen otros metodos como el historico o el de simulacion de Monte Carlo.



````python
#Ahora consideremos tres acciones cuyas rentabilidades anuales  mensuales esperadas son: 5% 6% y 7%
#Y cuya matriz de covarianzas es:
rentabilidades_esperadas = np.array([0.05, 0.06, 0.07])
covarianzas = np.array([[0.05, 0.07, 0.11],
                         [0.07, 0.10, 0.13],
                         [0.11, 0.13, 0.15]])
#Nos piden hallar VaR mensual al 97% de una cartera que destina 50% a la accion 1, 25% a la accion 2 y 25% a la accion 3
pesos = np.array([0.5, 0.25, 0.25])
#Luego apartado b. Si invirtiera 300000 euros en esa cartera cual seria el VaR mensual absoluto  al 97%
# c. Calcula la direccion y sentido en que se modificaria la cartera para que el VaR 97% disminuyera lo mas rapidamente posible
#d. Calcula los VaR marginales al 97% de cada una d elas 3 acciones
#Cambie la cartera original en 1% de forma que el var 97% disminuya lo mas rapidamente posible  y calcula VaR 97% de la nueva cartera

#a) rentabilidad mensual esperada
rentabilidad_mensual_esperada = np.dot(pesos, rentabilidades_esperadas)
print("Rentabilidad mensual esperada de la cartera:", rentabilidad_mensual_esperada)

#Hay alguna forma para calcular el sumatorio de la matriz de covarianzas con los pesos para hallar la varianza de la cartera
varianza_cartera = np.dot(pesos.T, np.dot(covarianzas, pesos))
#Como funciona np.dot? 
#Funciona como una multiplicacion de matrices. En este caso primero multiplica la matriz de covarianzas por el vector de pesos, y luego multiplica el resultado por la transpuesta del vector de pesos.
print("Varianza de la cartera:", varianza_cartera)
desviacion_estandar_cartera = np.sqrt(varianza_cartera)
print("Desviación estándar de la cartera:", desviacion_estandar_cartera)

#Phi ^-1 (0.03)
from scipy.stats import norm
z_97 = norm.ppf(0.03)
print("Valor z para VaR al 97%:", z_97)

VaR_97 = - (rentabilidad_mensual_esperada + z_97 * desviacion_estandar_cartera)
print("VaR mensual al 97% de la cartera (en proporción):", VaR_97)

#Por cada euro que invierto en la cartera lo minimo que puedo perder mensualmente si caigo en uno de de los  3% peores casos es 0504775509...neuros. 
#Si invierto 300000 euros en la cartera
inversion = 300000
VaR_97_absoluto = VaR_97 * inversion
print("VaR mensual al 97% de la cartera (absoluto para 300000 euros):", VaR_97_absoluto)

#c) Direccion y sentido para disminuir VaR 97% rapidamente. 
#Para su calculo se usan los VaR marginales de cada accion. Por medio de la forumula:
#(sumatorio de i=1 a n de D_i -n D_1 /n; sumatorio de i=1 a n de D_i -n D_2 /n; sumatorio de i=1 a n de D_i -n D_3 /n)
#Donde D_i es el VaR marginal de la accion i
VaR_marginales = np.zeros(3)
for i in range(3):
    pesos_perturbados = pesos.copy()
    pesos_perturbados[i] += 0.0001  # Pequeña perturbación
    varianza_perturbada = np.dot(pesos_perturbados.T, np.dot(covarianzas, pesos_perturbados))
    desviacion_estandar_perturbada = np.sqrt(varianza_perturbada)
    VaR_perturbado = - (np.dot(pesos_perturbados, rentabilidades_esperadas) + z_97 * desviacion_estandar_perturbada)
    VaR_marginales[i] = (VaR_perturbado - VaR_97) / 0.0001  # Derivada aproximada       
print("VaR marginales de cada acción:", VaR_marginales) 
direccion_sentido = (np.sum(VaR_marginales) - 3 * VaR_marginales) / 3
print("Dirección y sentido para disminuir VaR 97% rápidamente:", direccion_sentido)     

#d) Cambiamos la cartera original en un 1% en la direccion y sentido calculados
nuevo_pesos = pesos + 0.01 * direccion_sentido
#Normalizamos los pesos para que sumen 1
nuevo_pesos /= np.sum(nuevo_pesos)
print("Nuevos pesos de la cartera después del cambio:", nuevo_pesos)
#Calculamos el nuevo VaR 97% con los nuevos pesos
varianza_nueva_cartera = np.dot(nuevo_pesos.T, np.dot(covarianzas, nuevo_pesos))
desviacion_estandar_nueva_cartera = np.sqrt(varianza_nueva_cartera)
VaR_97_nuevo = - (np.dot(nuevo_pesos, rentabilidades_esperadas) + z_97 * desviacion_estandar_nueva_cartera)
print("Nuevo VaR mensual al 97% de la cartera después del cambio (en proporción):", VaR_97_nuevo)               
#El nuevo VaR 97% es menor que el anterior, por lo que hemos conseguido disminuir el VaR 97% de la cartera.
````

````python
# Todo lo aprendido anteriormente vamos a intentar aplicarlo a datos reales
try:
    import yfinance as yf
except ImportError as e:
    raise ImportError("Falta instalar yfinance. Ejecuta: pip install yfinance") from e

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import date, timedelta

# ¿Qué acciones son las más rentables en los últimos 2 años?
# Universo: 20 grandes empresas europeas (tickers de Yahoo Finance).
# Nota: algunos tickers pueden fallar si Yahoo los cambia; si pasa, quítalos o corrígelos y re-ejecuta.
tickers = [
    # Países Bajos / AS
    "ASML.AS", "SHEL.AS",
    # Alemania / DE
    "SAP.DE", "SIE.DE", "ALV.DE",
    # Francia / PA
    "MC.PA", "OR.PA", "TTE.PA", "SAN.PA", "RMS.PA", "AI.PA",
    # Suiza / SW
    "NESN.SW", "NOVN.SW", "ROG.SW",
    # Reino Unido / L
    "AZN.L", "ULVR.L", "HSBA.L",
    # Dinamarca / CO
    "NOVO-B.CO",
    # España / MC
    "ITX.MC",
    # Italia / MI
    "ENEL.MI",
]

# Ventana temporal (años)
window_years = 2

end = date.today()
start = end - timedelta(days=int(365.25 * window_years) + 10)  # margen extra
prices_raw = yf.download(
    tickers=tickers,
    start=start,
    end=end,
    auto_adjust=True,
    progress=False,
    threads=True,
    group_by="column",
 )

if prices_raw.empty:
    raise RuntimeError("No se han descargado datos. Revisa conexión a internet o los tickers.")

# Con auto_adjust=True, la serie ajustada es la columna Close
prices = prices_raw["Close"]
if isinstance(prices, pd.Series):
    prices = prices.to_frame()

# Limpieza básica
prices = prices.sort_index().ffill().dropna(axis=1, how="all")
prices = prices.dropna(axis=0, how="all")

# Rentabilidad total y CAGR (anualizada)
first = prices.iloc[0]
last = prices.iloc[-1]
total_return = (last / first) - 1

years = (prices.index[-1] - prices.index[0]).days / 365.25
cagr = (1 + total_return) ** (1 / years) - 1

ranking = (
    pd.DataFrame({"TotalReturn": total_return, "CAGR": cagr})
    .sort_values("CAGR", ascending=False)
    .dropna()
 )

print(f"Ventana usada: {prices.index[0].date()} → {prices.index[-1].date()} (~{years:.2f} años)")
print(f"Tickers con datos: {len(ranking)} / {len(tickers)}")
display(ranking.head(20).style.format({"TotalReturn": "{:.2%}", "CAGR": "{:.2%}"}))

# Estimación de rentabilidad media 'esperada' (histórica)
## En teoría de carteras (Markowitz), una aproximación común a E[R] es la media muestral de retornos.
## Ojo: es un estimador histórico (no garantiza futuro).
rets_d = prices.pct_change().dropna(how="all")

# Media diaria (aritmética)
mu_d = rets_d.mean()

# Annualización típica (252 días de mercado)
mu_ann_arith = mu_d * 252
mu_ann_comp = (1 + mu_d) ** 252 - 1  # compuesta a partir de la media diaria

# Alternativa: media de log-retornos (más estable para componer)
logrets_d = np.log(prices / prices.shift(1)).dropna(how="all")
mu_log_ann = logrets_d.mean() * 252
mu_ann_from_log = np.exp(mu_log_ann) - 1

mu_table = (
    pd.DataFrame(
        {
            "MeanDaily": mu_d,
            "MeanAnn(Arith≈)": mu_ann_arith,
            "MeanAnn(Comp)": mu_ann_comp,
            "MeanAnn(Log)": mu_ann_from_log,
        }
    )
    .dropna()
)

# Muestra solo los tickers con datos válidos en el ranking
mu_table = mu_table.loc[ranking.index]

display(
    mu_table.sort_values("MeanAnn(Comp)", ascending=False)
    .style.format(
        {
            "MeanDaily": "{:.3%}",
            "MeanAnn(Arith≈)": "{:.2%}",
            "MeanAnn(Comp)": "{:.2%}",
            "MeanAnn(Log)": "{:.2%}",
        }
    )
)

# Gráfico: crecimiento de 1€ en el TOP 5 por CAGR
top = ranking.head(5).index.tolist()
growth = prices[top] / prices[top].iloc[0]
ax = growth.plot(figsize=(10, 5), linewidth=2)
ax.set_title(f"Crecimiento de 1€ (precios ajustados) — TOP 5 últimos ~{window_years} años")
ax.set_xlabel("Fecha")
ax.set_ylabel("Crecimiento")
ax.grid(True, alpha=0.3)
plt.show()

#HAGAMOS LO MISMO PARA EMPRESAS ESTADOUNIDENSES
# Universo: 20 grandes empresas estadounidenses (tickers de Yahoo Finance).
tickers_us = [  
    "AAPL", "MSFT", "GOOGL", "AMZN", "FB",
    "TSLA", "BRK-B", "JPM", "JNJ", "V",
    "WMT", "PG", "NVDA", "DIS", "MA",
    "HD", "PYPL", "BAC", "VZ", "ADBE",
]
# Ventana temporal (años)

window_years_us = 4
end_us = date.today()
start_us = end_us - timedelta(days=int(365.25 * window_years_us) + 10)  # margen extra
prices_raw_us = yf.download(        
    tickers=tickers_us,
    start=start_us,
    end=end_us,
    auto_adjust=True,
    progress=False,
    threads=True,
    group_by="column",
 )  
if prices_raw_us.empty:
    raise RuntimeError("No se han descargado datos. Revisa conexión a internet o los tickers.")
# Con auto_adjust=True, la serie ajustada es la columna Close
prices_us = prices_raw_us["Close"]
if isinstance(prices_us, pd.Series):
    prices_us = prices_us.to_frame()
# Limpieza básica
prices_us = prices_us.sort_index().ffill().dropna(axis=1, how="all")
prices_us = prices_us.dropna(axis=0, how="all")
# Rentabilidad total y CAGR (anualizada)
first_us = prices_us.iloc[0]
last_us = prices_us.iloc[-1]
total_return_us = (last_us / first_us) - 1
years_us = (prices_us.index[-1] - prices_us.index[0]).days / 365.25
cagr_us = (1 + total_return_us) ** (1 / years_us) - 1
ranking_us = (
    pd.DataFrame({"TotalReturn": total_return_us, "CAGR": cagr_us})
    .sort_values("CAGR", ascending=False)
    .dropna()
 )
print(f"Ventana usada: {prices_us.index[0].date()} → {prices_us.index[-1].date()} (~{years_us:.2f} años)")
print(f"Tickers con datos: {len(ranking_us)} / {len(tickers_us)}")
display(ranking_us.head(20).style.format({"TotalReturn": "{:.2%}", "CAGR": "{:.2%}"}))
# Estimación de rentabilidad media 'esperada' (histórica)
rets_d_us = prices_us.pct_change().dropna(how="all")
# Media diaria (aritmética)
mu_d_us = rets_d_us.mean()
# Annualización típica (252 días de mercado)
mu_ann_arith_us = mu_d_us * 252
mu_ann_comp_us = (1 + mu_d_us) ** 252 - 1  # compuesta a partir de la media diaria
# Alternativa: media de log-retornos (más estable para componer)
logrets_d_us = np.log(prices_us / prices_us.shift(1)).dropna(how="all" ) 
mu_log_ann_us = logrets_d_us.mean() * 252
mu_ann_from_log_us = np.exp(mu_log_ann_us) - 1
mu_table_us = (
    pd.DataFrame(
        {
            "MeanDaily": mu_d_us,
            "MeanAnn(Arith≈)": mu_ann_arith_us,
            "MeanAnn(Comp)": mu_ann_comp_us,
            "MeanAnn(Log)": mu_ann_from_log_us,
        }
    )
    .dropna()
)
# Muestra solo los tickers con datos válidos en el ranking
mu_table_us = mu_table_us.loc[ranking_us.index]
display(
    mu_table_us.sort_values("MeanAnn(Comp)", ascending=False)
    .style.format(
        {
            "MeanDaily": "{:.3%}",
            "MeanAnn(Arith≈)": "{:.2%}",
            "MeanAnn(Comp)": "{:.2%}",
            "MeanAnn(Log)": "{:.2%}",
        }
    )
)
# Gráfico: crecimiento de 1€ en el TOP 5 por CAGR
top_us = ranking_us.head(5).index.tolist()
growth_us = prices_us[top_us] / prices_us[top_us].iloc[0]
ax_us = growth_us.plot(figsize=(10, 5), linewidth=2)
ax_us.set_title(f"Crecimiento de 1$ (precios ajustados) — TOP 5 últimos ~{window_years_us} años")
ax_us.set_xlabel("Fecha")
ax_us.set_ylabel("Crecimiento")
ax_us.grid(True, alpha=0.3)
plt.show()
#Hemos aplicado todo lo aprendido a datos reales de mercado financiero tanto europeo como estadounidense.
````

````python
# Comparar mercados/sectores y comprobar diversificación (enfoque Markowitz)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Pandas 3.x: "M" ya no vale; usa "ME" (month-end)
MONTH_FREQ = "ME"

# 1) Define tu universo (puedes cambiar/añadir tickers)
## Consejo práctico: usa universos por "bloques" (regiones/sectores) y luego amplía a acciones concretas.
universe_groups = {
    # Índices / regiones (ETFs como proxy)
    "US Broad (SPY)": ["SPY"],
    "US Tech (QQQ)": ["QQQ"],
    "Europe (IEUR)": ["IEUR"],
    "Japan (EWJ)": ["EWJ"],
    "Emerging (EEM)": ["EEM"],
    "World (VT)": ["VT"],
    # Sectores USA (ETFs sectoriales, proxies)
    "Tech (XLK)": ["XLK"],
    "Healthcare (XLV)": ["XLV"],
    "Financials (XLF)": ["XLF"],
    "Energy (XLE)": ["XLE"],
    "Industrials (XLI)": ["XLI"],
    "Consumer Staples (XLP)": ["XLP"],
    # Ejemplo de acciones (puedes sustituir por las que quieras analizar)
    "MegaCap US (ejemplo)": ["MSFT", "AAPL", "NVDA", "AMZN", "GOOGL", "META"],
}

# Aplanar tickers únicos manteniendo orden aproximado
all_tickers = []
for _g, _ts in universe_groups.items():
    for t in _ts:
        if t not in all_tickers:
            all_tickers.append(t)

# 2) Descarga precios ajustados y calcula retornos mensuales (más robusto para Markowitz que diario)
end = pd.Timestamp.today().normalize()
start = end - pd.Timedelta(days=int(365.25 * window_years) + 30)

prices_all_raw = yf.download(
    tickers=all_tickers,
    start=start.date(),
    end=end.date(),
    auto_adjust=True,
    progress=False,
    threads=True,
    group_by="column",
 )

if prices_all_raw.empty:
    raise RuntimeError("No se han descargado datos del universo. Revisa conexión o tickers.")

prices_all = prices_all_raw["Close"]
if isinstance(prices_all, pd.Series):
    prices_all = prices_all.to_frame()

prices_all = prices_all.sort_index().ffill().dropna(axis=1, how="all")
prices_m = prices_all.resample(MONTH_FREQ).last().dropna(how="all")
rets_m = prices_m.pct_change().dropna(how="all")

# 3) Estimación Markowitz: E[R] y Sigma (annualizadas desde mensual)
mu_m = rets_m.mean()
mu_ann = (1 + mu_m) ** 12 - 1  # anual compuesta
Sigma_ann = rets_m.cov() * 12   # cov anualizada aproximada

universe = mu_ann.index.tolist()
mu_vec = mu_ann.values
Sigma = Sigma_ann.loc[universe, universe].values

print(f"Universo con datos: {len(universe)} activos")
print(f"Ventana mensual: {rets_m.index[0].date()} → {rets_m.index[-1].date()} ({len(rets_m)} meses)")

display(
    pd.DataFrame({"E[R] anual (hist)": mu_ann}).sort_values("E[R] anual (hist)", ascending=False)
    .style.format({"E[R] anual (hist)": "{:.2%}"})
 )

# 4) Optimización simple: máximo Sharpe (sin cortos) + métricas de diversificación
rf = 0.02          # tipo libre de riesgo anual (ajústalo)
max_weight = 0.25  # límite por activo para evitar concentración
eps = 1e-12

def portfolio_stats(w, mu, Sigma):
    w = np.asarray(w, dtype=float)
    ret = float(w @ mu)
    vol = float(np.sqrt(max(w @ Sigma @ w, 0.0)))
    sharpe = (ret - rf) / (vol + eps)
    return ret, vol, sharpe

def neg_sharpe(w, mu, Sigma):
    return -portfolio_stats(w, mu, Sigma)[2]

n = len(universe)
w0 = np.ones(n) / n
bounds = [(0.0, max_weight) for _ in range(n)]
cons = ({"type": "eq", "fun": lambda w: np.sum(w) - 1.0},)

res = minimize(neg_sharpe, w0, args=(mu_vec, Sigma), method="SLSQP", bounds=bounds, constraints=cons)
if not res.success:
    print("Aviso: la optimización no convergió ->", res.message)

w_opt = np.clip(res.x, 0.0, None)
w_opt = w_opt / w_opt.sum()  # normaliza por seguridad
ret_opt, vol_opt, sh_opt = portfolio_stats(w_opt, mu_vec, Sigma)

# Métricas de diversificación
hhi = float(np.sum(w_opt ** 2))
eff_n = float(1.0 / hhi) if hhi > 0 else np.nan
max_w = float(w_opt.max())

weights = pd.Series(w_opt, index=universe).sort_values(ascending=False)
print(f"Cartera (max Sharpe) — E[R]: {ret_opt:.2%} | Vol: {vol_opt:.2%} | Sharpe: {sh_opt:.2f}")
print(f"Concentración — max peso: {max_w:.2%} | HHI: {hhi:.3f} | Nº efectivo activos: {eff_n:.2f}")

display(weights.head(20).to_frame("Peso").style.format({"Peso": "{:.2%}"}))

# Correlaciones (para idea de diversificación)
corr = rets_m[universe].corr()
avg_corr = float((corr.values[np.triu_indices_from(corr.values, k=1)]).mean())
print(f"Correlación media (pares): {avg_corr:.3f}")

# Gráfico pesos
ax = weights.head(15).iloc[::-1].plot(kind="barh", figsize=(8, 5))
ax.set_title("Top pesos — cartera max Sharpe (histórica)")
ax.set_xlabel("Peso")
ax.grid(True, axis="x", alpha=0.3)
plt.show()

# 5) Comparación por grupos (mercados/sectores) usando proxies (ETFs)
group_rows = []
for g, ts in universe_groups.items():
    ts_ok = [t for t in ts if t in prices_all.columns]
    if not ts_ok:
        continue
    # si el grupo tiene varios tickers (p.ej. MegaCaps), igual-peso dentro del grupo
    p = prices_all[ts_ok].dropna(how="all")
    if p.empty:
        continue
    p_m = p.resample(MONTH_FREQ).last().dropna(how="all")
    r_m = p_m.pct_change().dropna(how="all")
    if r_m.empty:
        continue
    r_eq = r_m.mean(axis=1)
    mu_g = r_eq.mean()
    er_ann = (1 + mu_g) ** 12 - 1
    vol_ann = float(r_eq.std(ddof=1) * np.sqrt(12))
    sharpe_g = (er_ann - rf) / (vol_ann + eps)
    group_rows.append((g, er_ann, vol_ann, sharpe_g, len(ts_ok)))

groups_df = pd.DataFrame(group_rows, columns=["Grupo", "E[R] anual", "Vol anual", "Sharpe", "n"]).set_index("Grupo")
display(groups_df.sort_values("Sharpe", ascending=False).style.format({"E[R] anual": "{:.2%}", "Vol anual": "{:.2%}", "Sharpe": "{:.2f}"}))

print("Nota: esto es rendimiento histórico del periodo; no implica que el futuro sea igual.")
````

````python
#Llamamos a tickers de 3 acciones de EEUU de las que dan mas rentabilidad. tickers = ["AAPL", "MSFT", "NVDA"]

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from datetime import date, timedelta

tickers = ["AAPL", "MSFT", "TSLA"]
data = yf.download(  tickers, start="2020-01-01", end="2025-01-01")["Close"]
ret = data.pct_change().dropna()

pesos = np.array([0.4, 0.4, 0.2])
port_ret = (ret * pesos).sum(axis=1)

prom_annual = port_ret.mean() * 252
std_annual = port_ret.std() * np.sqrt(252)

sharpe = (prom_annual ) / std_annual
    
print(f"Sharpe =  {sharpe:.4f}")
    
#Graficame la cartera eficiente de markowitz con esas 3 acciones
def cartera_eficiente(mu, sigma, num_portfolios=10000):
    results = np.zeros((3, num_portfolios))
    for i in range(num_portfolios):
        weights = np.random.random(len(mu))
        weights /= np.sum(weights)
        port_return = np.dot(weights, mu)
        port_std = np.sqrt(np.dot(weights.T, np.dot(sigma, weights)))
        results[0,i] = port_return
        results[1,i] = port_std
        results[2,i] = (port_return) / port_std  # Sharpe ratio sin rf
    return results
mu = ret.mean() * 252
sigma = ret.cov() * 252
results = cartera_eficiente(mu, sigma)
plt.figure(figsize=(10,6))
plt.scatter(results[1,:], results[0,:], c=results[2,:], cmap='viridis', marker='o', s=10, alpha=0.3)
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Riesgo (Desviación Estándar Anualizada)')
plt.ylabel('Rentabilidad Anualizada')
plt.title('Cartera Eficiente de Markowitz')
plt.show()

#Ahora dado un ALR como bonos USA a 10 años del 3% . Dibuja la CLM de la cartera eficiente de tal manera que corte con la cartera de mercado
rf = 0.03
def capital_market_line(rf, sharpe, sigma):
    return rf + sharpe * sigma
sigma_range = np.linspace(0, 0.5, 100)
cml = capital_market_line(rf, sharpe, sigma_range)
plt.figure(figsize=(10,6))
plt.plot(sigma_range, cml, color='red', label='Capital Market Line (CML)')
plt.scatter(results[1,:], results[0,:], c=results[2,:], cmap='viridis', marker='o', s=10, alpha=0.3)
plt.colorbar(label='Sharpe Ratio')      
plt.xlabel('Riesgo (Desviación Estándar Anualizada)')
plt.ylabel('Rentabilidad Anualizada')
plt.title('Cartera Eficiente de Markowitz con CML')
plt.legend()
plt.show()

#Cual es la cartera de mercado?
#La cartera de mercado es el punto donde la CML es tangente a la frontera eficiente.
#Calculamos el punto de tangencia
tangency_idx = np.argmax(results[2,:])
tangency_return = results[0, tangency_idx]
tangency_risk = results[1, tangency_idx]
print(f"Cartera de Mercado - Rentabilidad: {tangency_return:.2%}, Riesgo: {tangency_risk:.2%}")
#Cuales son los pesos de esa cartera de mercado?
#Para encontrar los pesos de la cartera de mercado, necesitamos resolver el problema de optimización para encontrar la cartera que maximiza el Sharpe ratio. Esto se puede hacer usando la función de optimización de scipy.
from scipy.optimize import minimize 
def neg_sharpe_ratio(weights, mu, sigma, rf):
    port_return = np.dot(weights, mu)
    port_std = np.sqrt(np.dot(weights.T, np.dot(sigma, weights)))
    return - (port_return - rf) / port_std
n = len(mu)
initial_weights = np.ones(n) / n
bounds = [(0, 1) for _ in range(n)]
constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
result = minimize(neg_sharpe_ratio, initial_weights, args=(mu.values, sigma .values, rf), method='SLSQP', bounds=bounds, constraints=constraints)
market_weights = result.x

print("Pesos de la Cartera de Mercado:")
for ticker, weight in zip(tickers, market_weights):
    print(f"{ticker}: {weight:.2%}")        


#La suma de los pesos es 1 o queda algo para el ALR?
suma_pesos = np.sum(market_weights)
print(f"Suma de los pesos de la cartera de mercado: {suma_pesos:.2f}")
#Pero deberia quedar algo para el ALR ya que el Sharpe ratio no es infinito
````
